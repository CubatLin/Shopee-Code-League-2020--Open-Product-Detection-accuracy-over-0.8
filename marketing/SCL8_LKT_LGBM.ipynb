### **欄位說明**

- country_code: 國家代碼.

- grass_date: email被寄送的日期.

- user_id: 人的key值

- subject_line_length: 信件的長度

- last_[open\/ login\/ checkout]_day: User最後打開Email/登入/買過商品的日期

- open_count_last_[10\/ 30/60]_days:過去30天打開email的數量.

- login_count_last_[10\/ 30/60]_days:user在N天登入的次數.

- checkout_count_last_[10\/ 30/60]_days:User在過去幾天購買的次數.

- open_flag: Y 目標變數.



### **Summary**

- 用ccountry_code跟domain建basic model.

- grass_date: email被寄送的日期(要被弄成weekday).

- Sorting age into age range.

- subject_line_length/attr 需要更多EDA.

- last_open_day most useful in all day_count.

#### 8/4 Ethan補上Cycle Feature

```
train_df['grass_date'] = pd.to_datetime(train_df['grass_date'], format='%Y-%m-%d %H:%M:%S')

#
day=[]
for date in train_df['grass_date']:
  day.append(date.day)

train_df['day']=day

#
import math
#cycle feature
train_df['day_ArcLen_tan'] = np.tan((2*math.pi/30)*(train_df['day']-1))#依tan(弧長)計算
```

##### Join User Info.

```
## 併入user資訊
train_df = train_df.merge(user_df, how='left', on='user_id')
```

```
train_df.columns
```

##### 用客戶屬性當grouping,計算60天內開信的CV

```
#Groupby 客戶屬性
log60_df = train_df.groupby(['attr_3'])['login_count_last_60_days'].agg(["var"]).reset_index()
log30_df = train_df.groupby(['attr_3'])['login_count_last_30_days'].agg(["var"]).reset_index()

log60_df.columns=['attr_3','log_60_var']
log30_df.columns=['attr_3','log_30_var']
```

```
train_df = train_df.merge(log30_df,on = 'attr_3',how = 'left').merge(log60_df,on = 'attr_3',how = 'left')
```

```
train_df['log60_cv'] = train_df['login_count_last_60_days']/train_df['log_60_var']
train_df['log30_cv'] = train_df['login_count_last_30_days']/train_df['log_30_var']
```

```
#Groupby 客戶屬性
open60_df = train_df.groupby(['attr_3'])['open_count_last_60_days'].agg(["var"]).reset_index()
open10_df = train_df.groupby(['attr_3'])['open_count_last_10_days'].agg(["var"]).reset_index()

open60_df.columns=['attr_3','open_60_var']
open10_df.columns=['attr_3','open_10_var']
```

```
train_df = train_df.merge(open60_df,on = 'attr_3',how = 'left').merge(open10_df,on = 'attr_3',how = 'left')
```

```
train_df['open60_cv'] = train_df['open_count_last_60_days']/train_df['open_60_var']
train_df['open10_cv'] = train_df['open_count_last_10_days']/train_df['open_10_var']
```

##### 把FI高的變數做變化

```
#關鍵一波 weekday
train_df['grass_date'] = pd.to_datetime(train_df['grass_date'])
train_df['weekday'] = train_df['grass_date'].dt.dayofweek
```

```
train_df['sub_last']=train_df['subject_line_length']*train_df['last_open_day']
train_df['sub_open60']=train_df['subject_line_length']*train_df['open_count_last_60_days']
train_df['open60_last']=train_df['open_count_last_60_days']*train_df['last_open_day']
train_df['open60/open10'] = train_df['open_count_last_60_days']/train_df['open_count_last_10_days']

train_df['age_ArcLen'] = train_df['age']*train_df['day_ArcLen_tan']
train_df['open60/open10'] = train_df['open_count_last_60_days']/train_df['open_count_last_10_days']
train_df['age_day']=train_df['age']*train_df['day']
```

```
train_df['sub_weekday']=train_df['subject_line_length']*train_df['weekday']
train_df['last_weekday']=train_df['last_open_day']*train_df['weekday']
train_df['open60_weekday']=train_df['open60_last']*train_df['weekday']
train_df['age_weekday']=train_df['age']*train_df['weekday']
```

```
le = preprocessing.LabelEncoder()
le.fit(train_df['domain'])
train_df['domain'] = le.transform(train_df['domain'])
```

```
train_df = train_df.drop(['open_count_last_60_days', 'open_60_var', 'open_count_last_10_days','open_10_var',
                              'login_count_last_60_days','log_60_var','login_count_last_30_days',
                              'log_30_var','attr_1', 'attr_2','attr_3' 
                          ,'user_id'  ],axis=1)
```

## 抽離身心 交給LGB

```
train_data = train_df[train_df['open_flag']!=-1]
train_data = train_data.drop(['grass_date','row_id'],axis=1)

train_data.shape
```

```
train_data.columns
```

```
X_train, X_val= train_test_split(
    train_data,
    test_size = 0.15,
    stratify = train_data['open_flag']
)
```

```
X_train.columns
```

```
d_train=lgb.Dataset(X_train.drop(['open_flag'], axis=1), label=X_train['open_flag'])
d_val=lgb.Dataset(X_val.drop(['open_flag'], axis=1), label=X_val['open_flag'])
#setting up the parameters
params={}
params['learning_rate']=0.001#0.001
params['boosting_type']='gbdt' #GradientBoostingDecisionTree
params['objective']='binary' #Multi-class target feature
params['metric']='binary' #metric for multi-class
params['max_depth']=18
params['feature_fraction'] = 0.9 # sub_row
params['bagging_fraction'] = 0.9 # sub_row
params['bagging_freq'] = 15
params['num_leaves'] = 50        # num_leaf
params['min_data'] = 100         # min_data_in_leaf
params['device'] = 'gpu'         # min_data_in_leaf
watchlist = [d_val]
```

```
clfer = lgb.train(params, d_train, 20000
                  ,watchlist
                  ,categorical_feature = ['country_code', 'domain', 'weekday']
                  ,early_stopping_rounds=300)
```

```
val_pred_prob = clfer.predict(X_val.drop(['open_flag'], axis=1))
```

```
val_pred = np.where(val_pred_prob>0.296 ,1, 0)
```

```
metrics.matthews_corrcoef(X_val['open_flag'], val_pred)

#0.5572215945689837
```

#### **看一下model效能**

```
pre = metrics.precision_score(X_val['open_flag'], val_pred)
re = metrics.recall_score(X_val['open_flag'], val_pred)
f1 = metrics.f1_score(X_val['open_flag'], val_pred)
print('pre:',pre,'\n' 
      're:',re,'\n' 
      'f1:', f1)
```
